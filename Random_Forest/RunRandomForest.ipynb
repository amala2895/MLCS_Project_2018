```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor

# Read in data and display first 5 rows

#data_train = pd.read_csv('train_data.csv')
data_train = pd.read_csv('train_data.csv')

#data_test = pd.read_csv('test_data.csv')
#data_test = pd.read_csv('split_files_test/xaa')
#print(data_train['CompanyName'].unique())
#print("test data here")
#print(data_test['CompanyName'].unique())
#print(data_train.head(10))
```

```python
#get one hot encoded company names
#df_appended = data_train.append(data_test)

#data_train = data_train[:10000]

#data_train = data_train.replace('ANALOG DEVICES','UNKNOWN')
#data_train = data_train.replace('BEL FUSE INC','UNKNOWN')
#data_train = data_train.replace('BERKSHIRE HATHAWAY','UNKNOWN')

data_encoded = pd.get_dummies(data_train)

#Divide the data in train and test
data_encoded_train = data_encoded[:10000]
data_encoded_test = data_encoded[10001:]

data_encoded_train.fillna(0)
data_encoded_test.fillna(0)

print(data_encoded.count())
print(data_encoded_train.count(axis=0))
print(data_encoded_test.count(axis=0))
```

```python

#train_data = data[(data["month"] == 1)|(data["month"] == 2)]
#test_data=data[(data["month"] == 3)]

labels_train = np.array(data_encoded_train['ChangeInStockPrice'])
labels_test = np.array(data_encoded_test['ChangeInStockPrice'])

train_data = np.array(data_encoded_train.drop(columns=['ChangeInStockPrice']))
test_data = np.array(data_encoded_test.drop(columns=['ChangeInStockPrice']))
```

```python

# Instantiate model with 100 decision trees and 20 leaves
rf = RandomForestRegressor(n_estimators = 50, random_state = 20)

# Train the model on training data
rf.fit(train_data, labels_train);
```

```python
#test_data.fillna(0)
```

```python

#print(df1.head(10))
#print(df1['CompanyName'].Unique())


# Use the forest's predict method on the test data
predictions = rf.predict(test_data)

#print(predictions)
# Calculate the absolute errors
for i in range(400):
    if(labels_test[i] != 0):
        errors[i] = abs(predictions[i] - labels_test[i])/abs(labels_test[i])
    else:    
        errors[i] = abs(predictions[i] - labels_test[i])

# Print out the mean absolute error (mae)
print('Percent Error:', round(np.mean(errors), 2), 'degrees.')

```
