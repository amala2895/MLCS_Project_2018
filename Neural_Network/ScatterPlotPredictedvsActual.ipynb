```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
actual = np.loadtxt('/Users/Anshu/MLCS_Project_copy/actual.txt', dtype=float)
predicted = np.loadtxt('/Users/Anshu/MLCS_Project_copy/prediction.txt', dtype=float)
```

```python
x = range(5000)

fig = plt.figure()
ax1 = fig.add_subplot(111)

ax1.scatter(x, actual, s=7, c='b', marker="s", label='actual change in stock')
ax1.scatter(x,predicted, s=7, c='r', marker="o", label='predicted change in stock')
plt.legend(loc='upper left');
plt.show()
```
