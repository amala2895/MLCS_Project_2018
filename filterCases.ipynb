```python
#initializing case data 
COLUMN_NAMES=['caseid','year','date','month','category','sub-category','filename','quarter','day']
casedata = pd.DataFrame(columns=COLUMN_NAMES)

#metadata
with open('bb2topic.pkl', 'rb') as f:
    data = pickle.load(f)


with open('bb2genis.pkl', 'rb') as f:
    data1 = pickle.load(f)

reader = csv.reader(open('caseid_date.csv', 'r'))

  
```

```python
#Filtering cases under Category 6 and 7 and storing as pickle file

df = pd.DataFrame([data, data1]).T
df.columns = ['d{}'.format(i) for i, col in enumerate(df, 1)]
#print(df)
dff=df.loc[df['d2'].isin([6,7])]
dff.to_pickle('filtered.pkl')
```

```python
dff = pd.read_pickle('filtered.pkl')
print(dff)
```

```python
#To check the Topics under Category 6 and 7
mapp={}
for index, row in dff.iterrows():
    if row['d1'] in mapp:
        mapp[row['d1']]=mapp[row['d1']]+1
    else:
    
        mapp[row['d1']]=1
print("sub-categories")
print("Categories\tNo of Cases")
for i in mapp:
    print("{}\t{}".format(i,mapp[i]))
```

```python
#storing metadata in case data
caseid_date = {}

for row in reader:
        
    k, v = row
    caseid_date[k] = v

print(len(caseid_date))
i=0
for index, row in dff.iterrows():
    try:
        i=i+1
        print(i)
        
        date=caseid_date[index]
        
        month=date.split("/")[0]
        day=date.split("/")[1]
        m=int(month)
        if m<4:
            q=1
        elif m>=4 and m<7:
            q=2
        elif m>=7 and m<10:
            q=3
        else:
            q=4
        casedata = casedata.append({'caseid':index,'date':date,'month':month,'category':row['d2'],'sub-category':row['d1'],'quarter':q},ignore_index=True)  
    except:
        print("error")
        print(index)

casedata = casedata.set_index('caseid')
print(len(casedata))


casedata.to_pickle("casedata.pkl")
```

```python
#Sentence Files

folders = os.listdir("sentences_new")
allfiles = collections.defaultdict(list)

for f in folders:
    if(1):
        #if(f=="sent_2013"):
        pathname = os.path.join("sentences_new", f)
        print(f)
        try:
            files=os.listdir(pathname)
            print(len(files))
            allfiles[f]=files
        except:
            print("folder cant be opened")
            continue
```
