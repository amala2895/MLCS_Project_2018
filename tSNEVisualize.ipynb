```python
import pandas as pd
import numpy as np
from sklearn.manifold import TSNE
from ggplot import *

# Load training data

training_data = pd.read_pickle('/Users/Anshu/MLCS_Project/training_data_CaseCompanyStockChange.pkl');

X = training_data.drop('ChangeInStockPrice',axis=1)
X_4940 = X.head(4940)
X_2470 = X.head(2470)

y = training_data[['ChangeInStockPrice']]
y_4940 = y.head(4940)
y_2470 = y.head(2470)

print('X =')
print(X.head(5))

print('y = ')
print(y.head(5))
```

```python
tsne = TSNE(n_components=2, verbose=1, perplexity=40, n_iter=300)
```

```python
#4940 data points
wordVector_to_columns = pd.DataFrame(X_4940['Vector'].values.tolist())

#200 data points
wordVector_to_columns_2470 = pd.DataFrame(X_2470['Vector'].values.tolist())

wordVector_to_columns.head(10)
```

```python
company_vector = X_4940.drop('Vector',axis=1)

company_vector_2470 = X_2470.drop('Vector', axis=1)

company_vector.head(10)
```

```python

wordColumn_companyVector = pd.concat([wordVector_to_columns,company_vector], axis=1)

wordColumn_companyVector_2470 = pd.concat([wordVector_to_columns_2470,company_vector_2470], axis=1)


wordColumn_companyVector['labels'] = y_4940
wordColumn_companyVector_2470['labels'] = y_2470

wordColumn_companyVector_2470.head(10)
```

```python
tsne_results = tsne.fit_transform(wordColumn_companyVector.values)

tsne_results_2470 = tsne.fit_transform(wordColumn_companyVector_2470.values)
```

```python

df_tsne = wordColumn_companyVector.loc[:,:].copy()
df_tsne['x-tsne'] = tsne_results[:,0]
df_tsne['y-tsne'] = tsne_results[:,1]

chart = ggplot( df_tsne, aes(x='x-tsne', y='y-tsne', color='labels') ) + geom_point(size=70,alpha=0.1) + scale_fill_brewer(palette="Reds") + ggtitle("tSNE dimensions colored by digit")
chart
```

```python
df_tsne_2470 = wordColumn_companyVector_2470.loc[:,:].copy()
df_tsne_2470['x-tsne'] = tsne_results_2470[:,0]
df_tsne_2470['y-tsne'] = tsne_results_2470[:,1]

chart = ggplot( df_tsne_2470, aes(x='x-tsne', y='y-tsne', color='labels') ) + geom_point(size=70,alpha=0.1) + scale_fill_brewer(palette="Reds") + ggtitle("tSNE dimensions colored by digit")
chart
```
