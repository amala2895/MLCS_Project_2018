```python
import pickle
import pandas as pd
import os
import shutil
import collections
import csv
from IPython.core.display import HTML
```

```python
#Initial Preparation 
path = "/Users/amaladeshpande/desktop/mlcs_project_final/Filtered/"
for i in range(2001,2014):
    path1=path+str(i)
    os.makedirs(path1, exist_ok=True)
    

    
```

```python
#metadata
with open('bb2topic.pkl', 'rb') as f:
    data = pickle.load(f)


with open('bb2genis.pkl', 'rb') as f:
    data1 = pickle.load(f)

reader = csv.reader(open('caseid_date.csv', 'r'))
```

```python

#Filtering cases under Category 6 and 7 and storing as pickle file

df = pd.DataFrame([data, data1]).T
df.columns = ['d{}'.format(i) for i, col in enumerate(df, 1)]
#print(df)
dff=df.loc[df['d2'].isin([6,7])]
dff.to_pickle('filtered.pkl')
```

```python
#Sentence Files

folders = os.listdir("sentences")
allfiles = collections.defaultdict(list)

for f in folders:
    if(1):
        #if(f=="sent_2013"):
        pathname = os.path.join("sentences", f)
        print(f)
        try:
            files=os.listdir(pathname)
            print(len(files))
            allfiles[f]=files
        except:
            print("folder cant be opened")
            continue
```

```python

dff = pd.read_pickle('filtered.pkl')

caseid_date = {}

for row in reader:
        
    k, v = row
    caseid_date[k] = v
```

```python

#initializing case data 
COLUMN_NAMES=['CaseId','Year','Date','Month','filename','Day']
casedata = pd.DataFrame(columns=COLUMN_NAMES)

tc=0
files_present={}
caseid_filenames={}
caseid_year={}


for folder in allfiles:
    
    print(folder)
    files=allfiles[folder]
        
    print(len(files))

    for file in files:
        file1=file.split("_")[0]
        if file1 in dff.index and "MajOp" in file:

            if file1 in files_present:
                print(file1)
                print("duplicate")




            else:
                    
                files_present[file1]=folder
        
            date=caseid_date[file1]
        
            month=date.split("/")[0]
            day=date.split("/")[1]
            year=folder.split("_")[1]
            casedata = casedata.append({'CaseId':file1,'Date':date,'Month':month,"Year":year,"Day":day,"filename":file},ignore_index=True)           
            try:
                    
                    
                    
                dest="/Users/amaladeshpande/desktop/mlcs_project_final/Filtered/"+str(year)  

                source="/Users/amaladeshpande/desktop/mlcs_project_final/sentences/"+files_present[file1]+"/"+file



                shutil.copy(source, dest)
                        
            except:
                print("error")
                print(file)
                print(folder)

print(len(files_present))
print(len(casedata))

casedata
```
